```julia
#-----------------------------------------------------------
# Set up time step
#-----------------------------------------------------------
tmax = 100;
level = 13;
nt = 2^level + 1;
deltat = tmax / nt;
#-----------------------------------------------------------
# Set up initial condition (case 2 cores)
# N: number of cores
# n: number of stars
# r0: initial core positions
# v0: initial core velocities
# m: core mass
# rmin: core minimum radius
# rmax: core maximum radius
# clockwise: core rotation direction
#-----------------------------------------------------------
N = 2;
n = 2000;
r0 = [400 500 0; -400 -500 0];
v0 = [-38 -10 0; 38 10 0];
m = [8*10^5; 8*10^5];
rmin = [200; 200];
rmax = [500; 500];
clockwise = [false; false];
"$(3nt*n*N) элементов в массиве - пощади оперативку!"
```

```julia
include("galaxies.jl")
```

```julia
#-----------------------------------------------------------
# Compute results
#-----------------------------------------------------------
# БЕРЕГИ ОПЕРАТИВКУ!!!
t, r, s_r = particleSystem(tmax, level, N, n, r0, v0, m, rmin, rmax, clockwise);
```

```julia
it = 7800
R = 2e3
dr = 0.4R
pth = s"C:\Users\User\Desktop\Mycop\plots2"
scatter(s_r[1:n,1,it], s_r[1:n,2,it], marker=(2, 0.9), legend = false, xaxis=("X", (-R, R), -R:dr:R),
                                                                        yaxis=("Y", (-R, R), -R:dr:R),);
scatter!(s_r[n+1:2*n,1,it], s_r[n+1:2*n,2,it], marker=(2, 0.9));
scatter!(r[:,1,it], r[:,2,it], marker=(10, 0.9) )
```

```julia
for it in 1:50:length(t)
    scatter(s_r[1:n,1,it], s_r[1:n,2,it], marker=(2, 0.9), legend = false, xaxis=("X", (-R, R), -R:dr:R),
                                                                        yaxis=("Y", (-R, R), -R:dr:R),);
    scatter!(s_r[n+1:2*n,1,it], s_r[n+1:2*n,2,it], marker=(2, 0.9));
    scatter!(r[:,1,it], r[:,2,it], marker=(10, 0.9) )
    savefig("$pth/img$it.png")
end
```
