```julia
# Классическое решение
function Wdq(t::Array{Float64,1}; n_mods::Int64 = 10, 
        steps::Int64 = 1000, ΔE = 0.5, τ = 1512., Er₁ = 0.4, Er = 1.)
    
    kT = 0.025
    b = Er*kT
    Ω = 0.17
    Δ = 5e-3Er
    S = Er₁/Ω

    Q(t) = ΔE - Er + 2Er*exp(-t/τ)
    # подинтегральное выражение
    F(t, n) = exp(-0.25( Q(t) - (n-1)*Ω )^2 / b)
    # метод трапеции с запоминанием предыдущего значения
    # интеграл считаем не каждый раз от нуля, а на отрезках
    # на которые разбивали время
    function trapez(foo, resprev, a, b, N, n)
        h = (b - a)/N
        result = 0.5*(foo(a, n) + foo(b, n))
        result += sum(i-> foo(a + i*h, n), 1:N-1)
        result *= h
        result += resprev
    end
    
    ntgrl = zeros(n_mods) # массив нулей
    reslt = [0.]
    
    for i = 2:length(t)
        # считаем интегралы для каждого повторения моды отдельно
        for n = 1:n_mods
            ntgrl[n] = trapez(F, ntgrl[n], t[i-1], t[i], steps, n)
        end
        
        In = sum( n-> S^(n-1)/factorial(n-1)*real(ntgrl[n]), 1:n_mods )
        
        push!(reslt, In)
    end
    0.25Δ^2 * sqrt(π/b) * exp(-S) .* reslt
end
```

# Полином Эрмита

$$H_n(x) = \sum^{n/2}_{j=0}(-1)^j\frac{n!}{j!(n-2j)!}(2x)^{n-2j}$$

```julia
H(n, x) = sum( j-> (-1)^j*factorial(n)*(2x)^(n-2j)/factorial(j)/factorial(n-2j), 0:n÷2 )
```

```julia
@benchmark H(4, 20.)
```

```julia
# детерминант трехдиагональной матрицы
Htd(n::Int, x::Float64) = 2^(0.5n)*det( Tridiagonal(ones(n-1), fill(sqrt(2)*x,n), [1.0n-i for i=1:n-1] ) )
```

```julia
# С использованием Гауссовых квадратур
function W2(t::Array{Float64,1}; n::Int64 = 10, steps::Int64 = 1000, ΔE = 0.5, τ = 1512., Er₁ = 0., Er = 1.)
    
    kT = 0.025
    b = sqrt(Er*kT)
    Ω = 0.17
    Δ = 5e-3Er

    function trapez(foo, resprev, a, b, N)
        h = (b - a)/N
        result = 0.5*(foo(a) + foo(b))
        result += sum(i-> foo(a + i*h), 1:N-1)
        result *= h
        result += resprev
    end
    
    F(x, t₁) = exp( im*ΔE*x/b - Er₁/Ω*(1-exp(-im*x*Ω/b)) - 
        im*Er*τ*(1 - exp(-x/(b*τ)) - 2exp(-t₁/τ)*(1-exp(-x/(b*τ))) ) )
    
    Hzeros = fzeros(x-> Htd(n, x), 0, sqrt(2n) ) # нули полинома
    if n != 200
        ω = [ Htd(n-1, Hzeros[i])^(-2) for i in 1:n÷2 ]
        ω .*= BigInt(2)^(n-1)*factorial(BigInt(n))*sqrt(pi)/n^2
    else
        ω = w # для n=200 есть посчитанные в массиве
    end
    
    fun(t₁) = sum( i-> ω[i]*F(Hzeros[i], t₁), 1:n÷2 )
    
    reslt = [0. + im*0.]
    In = 0. + im*0.
    for i in 2:length(t)
        In = trapez(fun, In, t[i-1], t[i], steps)
        push!(reslt, In)
    end
    
    real.(reslt) * Δ^2 * 0.5/b
end
```
