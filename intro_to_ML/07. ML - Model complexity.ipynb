## Сложность модели

В последнем блокноте мы увидели, как настроить модель, добавив параметр. Благодаря этому мы смогли приспособить эту модель к определенным данным. Эта посадка была идеальной, насколько позволяли цифры. 

В следующем блокноте мы увидим, что по мере того, как мы расширяем наш набор данных, подгонка модели к нашим данным будет становится более сложной и результат будет менее идеальным. 

Во-первых, мы обнаружим, что можем добавить сложность в нашу модель, добавив больше параметров. Мы увидим, что для набора данных с двумя точками данных мы можем снова получить «идеальное» соответствие нашей модели, добавив второй параметр в нашу модель. Однако мы не можем просто добавлять параметры в нашу модель каждый раз, когда добавляем точку данных в наш набор данных, так как это приведет к явлению называемому **переобучением**. На изображении ниже показаны набор данных, близкий к линейному, и модели, которые показывают недостаток параметров, подгонку и перенасыщение, слева направо:

<img src="data/model_fitting.png" alt="Drawing" style="width: 800px;"/>


На первом изображении модель учитывает наклон, по которому попадают данные, но не смещение. На втором изображении модель учитывает наклон и смещение данных. Добавление этого второго параметра (смещение) к модели создает намного лучшее соответствие.

Однако мы можем представить, что модель может иметь слишком много параметров, и мы начинаем приспосабливать не только особенности данных высокого уровня, но и шум. Это оснащение изображено на третьем изображении.

Наша цель состоит в том, чтобы соответствовать данным, но избегать *чрезмерного* соответствия данных!
